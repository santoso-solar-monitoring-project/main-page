```python
from scipy import signal
from matplotlib import pyplot as plt
import numpy as np
fs = 60
lpass = signal.remez(20, fs * np.r_[0, .1, .2, .5], [1, 0], fs=fs)
freq, resp = signal.freqz(lpass)
```

```python
plt.semilogy(freq * 100/(2*np.pi), np.abs(resp), 'b-')
```

```python
SINC = lambda w: lambda n: w/np.pi * np.sinc(w/np.pi*n) # cutoff `w` in radians
```

```python
s = SINC(np.pi * .45)(np.arange(-50,50))
plt.plot(s)
```

```python
plt.plot(lpass-s)
```

```python
freq_s, resp_s = signal.freqz(s)
plt.semilogy(freq_s * 100/(2*np.pi), np.abs(resp_s), 'b-')
```
